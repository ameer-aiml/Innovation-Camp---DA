```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scatter plot
x = np.array([2,6,8,34,56,23])
y = np.array([5,7,23,45,67,26])
plt.scatter(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# Line plot
x = np.array([2,6,8,34,56,23])
y = np.array([5,7,23,45,67,26])
plt.plot(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.show()
```

```python
# Line and Scatter Combined plot
x = np.array([2,6,8,34,56,23])
y = np.array([5,7,23,45,67,26])
plt.plot(x,y,c='r')
plt.scatter(x,y,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y-Axis')
plt.title('Line and Scatter combined Plot')
plt.show()
```

```python
# Bar Graph
person = np.array(['Mr X','Mr Y','Mr Z'])
weight = np.array([56,78,34])
plt.bar(person,weight,color=['r','g','b'])
plt.xlabel('Person Name')
plt.ylabel('Weight of the Person')
plt.title('Bar Graph - Person vs Weight')
plt.show()
```

```python
# Pie Chart

cities = np.array(['Bangalore','Mumbai','Delhi','Kolkata','Pune'])
sales = np.array([34500,45200,45230,12940,23045])
plt.pie(sales,explode=[0,0,0,0.2,0],labels=cities,colors=['r','g','b','pink','c'],autopct='%.2f%%')
plt.show()
```
